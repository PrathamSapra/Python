# INFINITE LOOP

````python
count = 0

while count < 5:
  print("Count", count)
````

# WHILE LOOP

````python
count = 0

while count < 5:
  print("count is", count)
  count = count + 1
````

# NESTED FOR LOOP

````python
for i in range(3):
  for j in range(3):
    print(f"i: {i}, j:{j}")

````

Nested While Loop

````python
i = 1
while i <= 3:
  j = 1
  while j <= 3:
    print(f"i: {i}, j: {j}")
    j+=1
  i+=1
````
